```python
dic_2 = {"Igor": 10,
          "Marcos": 9,
          "Júlia": 5.75,
          "Clarissa": 5.75}
for key,value in dic_2.items():
    if value > 5.75 and value>8:
        print(key + " está aprovado(a) com notão ")
    elif value == 5.75:
        print(key + " passou raspando ")
    else:
        print(key + "reprovado")
```

```python
dic_2 = {"Igor": 10,
          "Marcos": 9,
          "Júlia": 5.75,
          "Clarissa": 5.75,
         "Leo":3,
         "Leticia":8
        }
for key,value in dic_2.items():
    if value > 5.75:
        if value>8:
             print(key + " está aprovado(a) com notão ")
        else: 
             print(key + " está aprovado(a)")
    elif value == 5.75:
        print(key + " passou raspando ")
    else:
        print(key + "reprovado")
        
```

```python
#loop em uma lista e acumulando valores específicos:
lista_c = [100,200,10,2,1,6,22,27]

lista_maior_50 =[]
for index,valor in enumerate(lista_c):
    if valor>50:
        print("indice da lista = " + str(index))
        lista_maior_50.append(valor)
print(lista_maior_50)
```

## Numpy

```python
#Usando a lógica com numpy
import numpy as np
#convertendo para numpy
arr = np.array(lista_c)
print(arr)
print(type(arr))

#Selecionando valores maiores que 50 - retorna um array booleano
print(arr>50)

# imprimir apenas valores maiores que 50:
print(arr[arr>50])
```

```python
for ii in arr:
    print(ii)
```

```python
# media, minima, máximo, desvio padrão, percentil, mediana...
print("Media = " + str(arr.mean()))
print("min = " + str(arr.min()))
print("max = " + str(arr.max()))
print("desvio = " + str(arr.std()))
print("Percentil90 = " + str(np.percentile(arr,90))) # força o numpy a usar a função
print("Percentil50 = " + str(np.percentile(arr,50)))
print("Mediana = " + str(np.median(arr)))
```

```python
# Array de numeros aleatórios
val_min = 80
val_max = 600

array_temperatura_aleatoria = (val_max-val_min)*np.random.rand(1000) + val_min
print(array_temperatura_aleatoria)
```

```python
#Cinética
fx = 200*np.exp(array_temperatura_aleatoria) + array_temperatura_aleatoria**(-0.2)
print(fx)
```

```python
# Figuras MATPLOTLIB
#importando o pacote
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.plot(array_temperatura_aleatoria,fx)
```

```python
fig,ax = plt.subplots(2)
ax[0].plot(fx)
ax[1].scatter(array_temperatura_aleatoria,fx)
ax[1].set_ylabel('fx')
```
