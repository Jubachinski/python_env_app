# Numpy

```python
#importando o pacote
import numpy as np

#-------------Geração de numeros aleatórios

#valores maximos e minimos
val_max,val_min = 10.,0.5
#Número de valores aleatórios
n = 1000
# Utilizando pacote numpy e função uniform que está dentro do subpacote random
array_aleatorio = np.random.uniform(val_min,val_max,n)
print(array_aleatorio)
#Utilizando pacote numpy e função randint que está dentro do subpacote random
array_aleatorio_inteiro = np.random.randint(val_min,val_max,n)
print(array_aleatorio_inteiro)
```

```python
#Verificando o tamanho da matriz arr_aleatorio
#shape
print(array_aleatorio.shape)
print(np.shape(array_aleatorio))
#size
print(array_aleatorio.size)
```

```python
#Utilizando o reshape para formatar um vetor em uma matriz 2 linhas e n colunas
mat = array_aleatorio.reshape(2,-1)
print(np.shape(mat))
print(mat)
print(mat.size)
# o colchetes duplos separa as colunas - 2 colunas com 50 valores
```

```python
#Multiplicação de matrizes 
arr_resultado = array_aleatorio*array_aleatorio_inteiro
print(arr_resultado)
#mesmo shape
```

```python
#multiplicação de arrays de diferente dimensões
multiplicador = np.random.randint(1,10,2)

array_resultado2 = []
for multi in multiplicador:
    print(multi)
    array_resultado2.append(array_aleatorio_inteiro*multi)
print(array_resultado2)

array_resultado2 = np.concatenate(array_resultado2)
print(array_resultado2.shape)

array_resultado3 = np.vstack(array_resultado2)
print(array_resultado3.shape)

array_resultado4 = np.hstack(array_resultado2)
print(array_resultado4.shape)

arr_resultado5 = array_resultado4.transpose()
print(arr_resultado5)
```

```python
# Estatística com numpy

#Mínimo considerando todas as dimensões
min_absoluto = array_resultado7.min()
print(min_absoluto)
#Minímo para cada ponto (lat,lon) condiserando todos os tempos
min_latlon = array_resultado7.min(axis=2)
max_latlon = array_resultado7.max(axis=2)
print(min_latlon)

# atribuindo um valor NaN dentro da matriz em algum lugar
array_resultado7[2,3,1] = np.nan
```

```python
# Inserindo um valor
df.loc[1,'VALOR'] = -1
print(df['VALOR'])
```

```python
df.loc[df['VALOR']>np.nanpercentile(df['VALOR'],50),'VALOR'] = np.nan
df.describe()
```
