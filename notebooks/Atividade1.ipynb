```python
# utilizando np.random em Torricelli - para queda livre
import numpy as np

velocidade_inicial = np.random.rand(100) * 20
altura = np.random.rand(100) * 50
gravidade = 9.8

velocidade_final = ((velocidade_inicial)**2 + (2*gravidade*altura))**(1/2)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.plot(velocidade_final,altura)
ax.set_ylabel('altura (m)')
ax.set_xlabel('velocidade_final (m/s)')
```

```python
fig,ax = plt.subplots(2)
ax[0].plot(velocidade_final)
ax[0].set_ylabel('velocidade_final')
ax[1].scatter(velocidade_final,altura)
ax[1].set_ylabel('velocidade_final')
ax[1].set_xlabel('altura')
```

```python
fig,ax = plt.subplots(2)
ax[0].scatter(altura, velocidade_final)
ax[1].scatter(velocidade_inicial, velocidade_final)
```

```python
# utilizando movimento harmônico simples

fase = np.random.rand(100) *(2*np.pi)
frequencia_angular = np.random.rand(100)
amplitude = np.random.rand(100)*10
tempo = np.arange(0,100)

posicao = amplitude*np.cos((frequencia_angular*tempo)+fase)
```

```python
fig,bx = plt.subplots()
bx.plot(tempo,posicao)
bx.set_ylabel('posicao (m)')
bx.set_xlabel('tempo (s)')
```

```python
fig,bx = plt.subplots(2)
bx[0].scatter(posicao, amplitude)
bx[1].scatter(frequencia_angular, posicao)
```

```python
#velocidade do MHS
velocidade = -(frequencia_angular*amplitude*np.sin(frequencia_angular*tempo + fase))
aceleracao = -(((frequencia_angular)**2)*amplitude*np.cos(frequencia_angular*tempo + fase))
```

```python
fig,cx = plt.subplots()
cx.plot(posicao,velocidade)
cx.set_ylabel('velocidade(m/s)')
cx.set_xlabel('posicao (m)')
```

```python
fig,cx = plt.subplots()
cx.plot(velocidade,aceleracao)
cx.set_ylabel('velocidade(m/s)')
cx.set_xlabel('aceleracao (m/s^2)')
```
